```python
# # #Q1.

# # Matplotlib is a popular data visualization library in Python used to create static, animated, and interactive visualizations 
# in Python. It is a 2D plotting library that provides an interface similar to that of MATLAB. Matplotlib can be used to visualize
# data in various forms such as line plots, scatter plots, bar plots, histograms, etc.

# # Here are five plots that can be plotted using the Pyplot module of Matplotlib:

# # Line Plot - This plot is used to visualize the relationship between two variables by plotting a line connecting the data points. It can be created using the plot() function in Matplotlib.

# # Scatter Plot - This plot is used to visualize the relationship between two variables by plotting points on a 2D plane. It can
# be created using the scatter() function in Matplotlib.

# # Bar Plot - This plot is used to visualize the frequency or count of categorical data using rectangular bars. It can be created 
# using the bar() function in Matplotlib.

# # Histogram - This plot is used to visualize the distribution of a continuous variable by dividing the data into a set of 
# intervals and plotting the frequency of data points falling within each interval. It can be created using the hist() function 
# in Matplotlib.

# # Pie Chart - This plot is used to visualize the proportion of each category in a categorical variable as a slice of a pie. It
# can be created using the pie() function in Matplotlib.
```

```python
# Q2.
""" A scatter plot is a type of data visualization in which data points are plotted on a two-dimensional plane. It is a graph in
which the values of two variables are plotted along the x-axis and y-axis, and a point is plotted at the intersection of these
values. Scatter plots are used to visualize the relationship between two variables and identify any patterns or trends in the 
data.

In a scatter plot, each point represents an observation in the data set. The position of the point on the x-axis and y-axis 
corresponds to the value of the two variables being plotted. Scatter plots can be used to identify whether the variables are 
positively or negatively correlated, or whether there is no correlation at all.

Scatter plots are commonly used in scientific research, engineering, and business to explore the relationship between variables
and to identify outliers or anomalies in the data. They are also useful for visualizing large data sets and identifying clusters
or patterns that may not be apparent in other types of plots. """

import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.scatter(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('This is title')
plt.show()
```

```python
#Q4.
"""A bar plot is a type of data visualization in which the frequency or count of categorical data is represented using rectangular
bars. It is a graph in which the height or length of each bar represents the value of the data being plotted. Bar plots are used 
to compare and visualize the relative frequency or count of different categories.

Bar plots can be created using vertical or horizontal bars, depending on the orientation of the graph. Vertical bar plots are 
commonly used for comparing the frequency of different categories, while horizontal bar plots are used for visualizing the rank
order of categories.

Bar plots are widely used in scientific research, business, and data analysis to represent data in a clear and easy-to-understand
manner. They are particularly useful when comparing the frequency or count of multiple categories or when visualizing data across 
different groups or subsets of a data set. Bar plots can also be used to identify patterns or trends in the data, such as changes 
in frequency or count over time or across different subgroups of a data set. """


import numpy as np
import matplotlib.pyplot as plt
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.bar(company,profit)
```

```python
#It show in horizontal bar horizontal
plt.barh(company,profit)
```

```python
#Q3.

""" The subplot() function in Matplotlib is used to create multiple plots or subplots in a single figure. It is useful when we 
want to compare multiple plots or visualize different aspects of a data set in a single figure.By using the subplot() function,
we can create a range of plots, including line plots, scatter plots, bar plots, histograms, and many others. We can also 
customize the size, color, and style of each subplot, as well as the spacing between subplots and the overall layout of the
figure.

Overall, the subplot() function is a powerful tool for creating complex and informative data visualizations that can help us to
better understand and analyze our data. """

import matplotlib.pyplot as plt
import numpy as np

# Create the data for the line plots
x1 = np.array([0, 1, 2, 3, 4, 5])

y1 = np.array([0, 100, 200, 300, 400, 500])

plt.subplot(2,2,1)

plt.title('Line-1')

plt.plot(x1,y1)

x2 = np.array([0, 1, 2, 3, 4, 5])

y2 = np.array([50, 20, 40, 20, 60, 70])

plt.subplot(2,2,2)

plt.title('Line-2')
plt.plot(x2,y2,c='black')

x3 = np.array([0, 1, 2, 3, 4, 5])

y3 = np.array([10, 20, 30, 40, 50, 60])

plt.subplot(2,2,3)

plt.title('Line-3')
plt.plot(x3,y3,c='red')

x4 = np.array([0, 1, 2, 3, 4, 5])

y4 = np.array([200, 350, 250, 550, 450, 150])

plt.subplot(2,2,4)


plt.title('Line-4')

plt.plot(x4,y4, c='orange')

# Adjust the spacing between subplots
plt.subplots_adjust(wspace=0.5, hspace=0.5)

plt.show()


```

```python
#Q5
""" A Box Plot is also known as Whisker plot is created to display the summary of the set of data values having properties like
minimum, first quartile, median, third quartile and maximum. In the box plot, a box is created from the first quartile to the 
third quartile, a vertical line is also there which goes through the box at the median. Here x-axis denotes the data to be
plotted while the y-axis shows the frequency distribution"""

"""  Box plots are used to identify the central tendency, dispersion, and skewness of a dataset, as well as to detect any outliers. 
They provide a clear and concise way to summarize and compare the distribution of different datasets, and are particularly useful
for large datasets.

Box plots are commonly used in data analysis, statistics, and scientific research to visualize and compare the distribution of
data across different groups, such as different experimental conditions or groups of study participants. They can also be used 
to identify potential problems with the data, such as data points that fall outside the whiskers or do not fit the expected 
distribution pattern. Overall, box plots are a powerful tool for exploratory data analysis and for communicating findings to 
others. """


import numpy as np
import matplotlib.pyplot as plt
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
plt.boxplot(box1)
```

```python
plt.boxplot(box2)
```

```python
plt.boxplot([box1,box2])
```
